```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 발전량 예측 10시 & 17시

```python
pred1 = pd.DataFrame()
pred2 = pd.DataFrame()
for day in range(17,31): ## 10/30
    fcst_date = f"2023-10-{day}"
    last_time = f"2023-10-{day+1}"
    gen_fcst_10 = pd.read_csv(f"gen_fcst_10_{fcst_date}.csv")
    gen_fcst_17 = pd.read_csv(f"gen_fcst_17_{fcst_date}.csv")

    date = []
    for i in range(24):
        date1, date2, date3 = [], [], []
        if i<9:
            date1.append(f'{fcst_date} 0{i+1}:00:00+09:00')
            date1 = date1*5
        elif i==23:
            date3.append(f'{last_time} 00:00:00+09:00')
            date3 = date3*5
        else:
            date2.append(f'{fcst_date} {i+1}:00:00+09:00')
            date2 = date2*5

        date = date+date1+date2+date3

    model_id = [0,1,2,3,4]*24
    round1 = [1]*5*24
    round2 = [2]*5*24

    gen_fcst_10_df = pd.DataFrame()
    gen_fcst_10_df['round'] = round1
    gen_fcst_10_df['time'] = date
    gen_fcst_10_df['model_id'] = model_id
    gen_fcst_10_df['amount'] = np.array(gen_fcst_10.iloc[:,1:]).flatten()

    pred1 = pd.concat([pred1,gen_fcst_10_df],ignore_index=True)

    gen_fcst_17_df = pd.DataFrame()
    gen_fcst_17_df['round'] = round2
    gen_fcst_17_df['time'] = date
    gen_fcst_17_df['model_id'] = model_id
    gen_fcst_17_df['amount'] = np.array(gen_fcst_17.iloc[:,1:]).flatten()

    pred2 = pd.concat([pred2,gen_fcst_17_df],ignore_index=True)
```

## 날씨 예측 10시 & 17시

```python
weather_forecast1 = pd.DataFrame()
weather_forecast2 = pd.DataFrame()
for day in range(17,31): ## 10/30까지
    fcst_date = f"2023-10-{day}"
    last_time = f"2023-10-{day+1}"
    weather_fcst_10 = pd.read_csv(f"weather_fcst_10_{fcst_date}.csv")
    weather_fcst_17 = pd.read_csv(f"weather_fcst_17_{fcst_date}.csv")

    date = []
    for i in range(24):
        if i<9:
            date.append(f'{fcst_date} 0{i+1}:00:00+09:00')
        elif i==23:
            date.append(f'{last_time} 00:00:00+09:00')
        else:
            date.append(f'{fcst_date} {i+1}:00:00+09:00')

    round1 = [1]*24
    round2 = [2]*24
    
    weather_fcst_10_df = pd.DataFrame()
    weather_fcst_10_df['round'] = round1
    weather_fcst_10_df['time'] = date
    weather_fcst_10_df = pd.concat([weather_fcst_10_df,weather_fcst_10.iloc[:,1:]],axis=1)

    weather_forecast1 = pd.concat([weather_forecast1, weather_fcst_10_df],ignore_index=True) 

    weather_fcst_17_df = pd.DataFrame()
    weather_fcst_17_df['round'] = round2
    weather_fcst_17_df['time'] = date
    weather_fcst_17_df = pd.concat([weather_fcst_17_df,weather_fcst_17.iloc[:,1:]],axis=1)

    weather_forecast2 = pd.concat([weather_forecast2, weather_fcst_17_df],ignore_index=True) 

```
